```python
# Import libraries and dependencies
import datetime, requests
import yfinance as yf
import pandas as pd
from pathlib import Path
```

```python
aapl = pd.read_csv(
    Path("AAPL_quarterly_financials.csv")
)
```

```python
aapl
```

```python
# Grab ticker
ticker = yf.Ticker("aapl")
```

```python
# Get summary of company
print(f"Market Cap : {ticker.info['marketCap']}\n")
print(f"Employees : {ticker.info['fullTimeEmployees']}\n")
print(f"Business Summary : \n{ticker.info['longBusinessSummary']}")
```

```python
# Find dividends and stock splits
ticker.actions
```

```python
# Get income Statements
income_stmt = ticker.quarterly_income_stmt
income_stmt
```

```python
# Put income statements into a dataframe and transpose 
income_stmt_df = pd.DataFrame(income_stmt).T
income_stmt_df
```

```python
# Find EPS
eps_df = income_stmt_df['Basic EPS']
eps_df
```

```python
# Find Revenue
revenue_df = income_stmt_df['Total Revenue']
revenue_df
```
